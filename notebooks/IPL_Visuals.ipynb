```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
from matplotlib import rcParams
from matplotlib.patches import FancyBboxPatch
import warnings
warnings.filterwarnings('ignore')

# ── Palette (muted, editorial) ──
C1  = '#4C6EF5'   # indigo blue   — primary
C2  = '#F76707'   # burnt orange  — secondary
C3  = '#0CA678'   # teal          — accent
C4  = '#AE3EC9'   # plum          — accent 2
BG  = '#F8F9FA'   # near-white
CARD= '#FFFFFF'
TX  = '#212529'   # dark text
TX2 = '#868E96'   # muted text
DIV = '#DEE2E6'   # divider

rcParams['font.family']      = 'DejaVu Sans'
rcParams['axes.spines.top']  = False
rcParams['axes.spines.right']= False
rcParams['axes.spines.left'] = False
rcParams['axes.spines.bottom']= False
rcParams['axes.grid']        = True
rcParams['grid.color']       = DIV
rcParams['grid.linewidth']   = 0.8
rcParams['axes.axisbelow']   = True
rcParams['figure.facecolor'] = BG
rcParams['axes.facecolor']   = CARD
rcParams['text.color']       = TX
rcParams['xtick.color']      = TX2
rcParams['ytick.color']      = TX2
rcParams['xtick.labelsize']  = 10
rcParams['ytick.labelsize']  = 10

print('Setup complete.')
```

## Chart 1 — Toss Impact on Match Outcome

```python
fig, ax = plt.subplots(figsize=(8, 5))
fig.patch.set_facecolor(BG)
ax.set_facecolor(CARD)

labels   = ['Won Toss', 'Lost Toss']
values   = [50.15, 49.85]
colors   = [C1, C2]
x        = np.arange(len(labels))
bar_w    = 0.42

bars = ax.bar(x, values, width=bar_w, color=colors,
              zorder=3, linewidth=0)

# Round top corners effect via overlay
for bar, val, col in zip(bars, values, colors):
    ax.text(bar.get_x() + bar.get_width()/2, val + 0.4,
            f'{val}%', ha='center', va='bottom',
            fontsize=15, fontweight='bold', color=col)

# 50% reference line
ax.axhline(50, color='#CED4DA', linewidth=1.4, linestyle='--', zorder=2)
ax.text(1.32, 50.3, '50% baseline', color=TX2, fontsize=9)

ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12, fontweight='600', color=TX)
ax.set_ylim(48, 52)
ax.set_yticks([48, 49, 50, 51, 52])
ax.set_yticklabels(['48%','49%','50%','51%','52%'])
ax.set_ylabel('Win Rate', fontsize=11, color=TX2, labelpad=10)

ax.set_title('Does Winning the Toss Matter?',
             fontsize=16, fontweight='bold', color=TX, pad=18, loc='left')
ax.text(0, -0.13, 'Win rate of toss winners vs toss losers across all IPL matches',
        transform=ax.transAxes, fontsize=9, color=TX2)

# Insight badge
fig.text(0.98, 0.04,
         'Difference: just 0.30% — virtually no advantage',
         ha='right', fontsize=9, color=TX2, style='italic')

plt.tight_layout()
plt.savefig('chart1_toss.png', dpi=180, bbox_inches='tight', facecolor=BG)
plt.show()
print('Saved: chart1_toss.png')
```

## Chart 2 — Average Runs per Phase: Winners vs Losers

```python
# ── Data ── replace with your actual values if different
phases  = ['Powerplay\n(Ov 1–6)', 'Middle Overs\n(Ov 7–15)', 'Death Overs\n(Ov 16–20)']
winners_runs = [57.95, 80.93, 49.92]
losers_runs  = [50.18, 73.40, 45.92]
gaps         = [w - l for w, l in zip(winners_runs, losers_runs)]

x     = np.arange(len(phases))
bar_w = 0.36

fig, ax = plt.subplots(figsize=(10, 5.5))
fig.patch.set_facecolor(BG)
ax.set_facecolor(CARD)

b1 = ax.bar(x - bar_w/2, winners_runs, width=bar_w,
            color=C1, label='Winning Team', zorder=3, linewidth=0)
b2 = ax.bar(x + bar_w/2, losers_runs,  width=bar_w,
            color=C2, label='Losing Team',  zorder=3, linewidth=0, alpha=0.85)

# Value labels
for bar in b1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
            f'{bar.get_height():.1f}', ha='center', va='bottom',
            fontsize=10, fontweight='bold', color=C1)
for bar in b2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
            f'{bar.get_height():.1f}', ha='center', va='bottom',
            fontsize=10, fontweight='bold', color=C2)

# Gap annotations between bar pairs
for i, (w, l, g) in enumerate(zip(winners_runs, losers_runs, gaps)):
    mid_x = x[i]
    mid_y = max(w, l) + 4.5
    ax.annotate('', xy=(x[i] - bar_w/2 + bar_w*0.1, max(w,l) + 1.5),
                xytext=(x[i] + bar_w/2 - bar_w*0.1, max(w,l) + 1.5),
                arrowprops=dict(arrowstyle='-', color=DIV, lw=1.5))
    ax.text(mid_x, mid_y, f'+{g:.1f} runs',
            ha='center', va='bottom', fontsize=9,
            color=TX2, fontweight='600',
            bbox=dict(boxstyle='round,pad=0.3', facecolor=BG, edgecolor=DIV, linewidth=0.8))

ax.set_xticks(x)
ax.set_xticklabels(phases, fontsize=11, fontweight='600', color=TX)
ax.set_ylabel('Average Runs', fontsize=11, color=TX2, labelpad=10)
ax.set_ylim(0, 105)
ax.set_title('Where Winning Teams Pull Ahead',
             fontsize=16, fontweight='bold', color=TX, pad=18, loc='left')
ax.text(0, -0.15, 'Average runs scored per phase — winning vs losing teams (all innings combined)',
        transform=ax.transAxes, fontsize=9, color=TX2)

legend = ax.legend(frameon=True, framealpha=1, edgecolor=DIV,
                   fontsize=10, loc='upper left',
                   facecolor=CARD)

plt.tight_layout()
plt.savefig('chart2_phase_runs.png', dpi=180, bbox_inches='tight', facecolor=BG)
plt.show()
print('Saved: chart2_phase_runs.png')
```

## Chart 3 — Phase Dominance Win Rate (Batting First vs Chasing)

```python
phases_short = ['Powerplay', 'Middle Overs', 'Death Overs']

bat_first_dom  = [67.3, 66.1, 53.2]
bat_first_lost = [30.1, 29.4, 48.0]
chasing_dom    = [70.8, 71.7, 50.0]
chasing_lost   = [32.9, 33.7, 47.6]

x     = np.arange(len(phases_short))
bar_w = 0.2

fig, axes = plt.subplots(1, 2, figsize=(14, 5.5), sharey=True)
fig.patch.set_facecolor(BG)

titles   = ['Batting First', 'Chasing']
dom_vals = [bat_first_dom,  chasing_dom]
lst_vals = [bat_first_lost, chasing_lost]

for ax, title, dom, lst in zip(axes, titles, dom_vals, lst_vals):
    ax.set_facecolor(CARD)

    b1 = ax.bar(x - bar_w/2, dom, width=bar_w,
                color=C1, label='Dominated Phase', zorder=3, linewidth=0)
    b2 = ax.bar(x + bar_w/2, lst, width=bar_w,
                color=C2, label='Lost Phase', zorder=3, linewidth=0, alpha=0.85)

    for bar in b1:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
                f'{bar.get_height():.1f}%', ha='center', va='bottom',
                fontsize=9, fontweight='bold', color=C1)
    for bar in b2:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
                f'{bar.get_height():.1f}%', ha='center', va='bottom',
                fontsize=9, fontweight='bold', color=C2)

    # 50% coin-flip line
    ax.axhline(50, color='#ADB5BD', linewidth=1.2, linestyle='--', zorder=2)
    ax.text(2.35, 51, 'Coin flip', color=TX2, fontsize=8)

    ax.set_xticks(x)
    ax.set_xticklabels(phases_short, fontsize=10, fontweight='600', color=TX)
    ax.set_ylim(0, 90)
    ax.set_title(title, fontsize=13, fontweight='bold', color=TX, pad=12)
    ax.legend(frameon=True, framealpha=1, edgecolor=DIV,
              fontsize=9, facecolor=CARD)

axes[0].set_ylabel('Win Rate (%)', fontsize=11, color=TX2, labelpad=10)

fig.suptitle('Which Phase Actually Decides the Match?',
             fontsize=17, fontweight='bold', color=TX, y=1.01, x=0.05, ha='left')
fig.text(0.05, -0.04,
         'Win rate when a team dominates (scores more runs in) each phase vs when they lose that phase',
         fontsize=9, color=TX2)

plt.tight_layout()
plt.savefig('chart3_phase_dominance.png', dpi=180, bbox_inches='tight', facecolor=BG)
plt.show()
print('Saved: chart3_phase_dominance.png')
```

## Chart 4 — Bowling Strategy Matrix (180+ vs Below 180)

```python
labels_matrix = [
    'Wickets +\nEconomical',
    'Wickets +\nExpensive',
    'No Wickets +\nEconomical',
    'No Wickets +\nExpensive'
]
vals_high = [86.9, 70.7, 74.6, 52.1]
vals_low  = [47.9, 48.7, 23.5, 11.2]

x     = np.arange(len(labels_matrix))
bar_w = 0.35

fig, ax = plt.subplots(figsize=(12, 5.5))
fig.patch.set_facecolor(BG)
ax.set_facecolor(CARD)

b1 = ax.bar(x - bar_w/2, vals_high, width=bar_w,
            color=C1, label='Defending 180+', zorder=3, linewidth=0)
b2 = ax.bar(x + bar_w/2, vals_low,  width=bar_w,
            color=C4, label='Defending Below 180', zorder=3, linewidth=0, alpha=0.85)

for bar in b1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
            f'{bar.get_height():.1f}%', ha='center', va='bottom',
            fontsize=10, fontweight='bold', color=C1)
for bar in b2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8,
            f'{bar.get_height():.1f}%', ha='center', va='bottom',
            fontsize=10, fontweight='bold', color=C4)

ax.axhline(50, color='#ADB5BD', linewidth=1.2, linestyle='--', zorder=2)
ax.text(3.52, 51.5, '50%', color=TX2, fontsize=8)

ax.set_xticks(x)
ax.set_xticklabels(labels_matrix, fontsize=10, fontweight='600', color=TX)
ax.set_ylabel('% Matches Defended Successfully', fontsize=11, color=TX2, labelpad=10)
ax.set_ylim(0, 105)
ax.set_title('Bowling Strategy Flips Based on Total Defended',
             fontsize=16, fontweight='bold', color=TX, pad=18, loc='left')
ax.text(0, -0.16,
        'When defending 180+, economy alone wins 74.6% — but below 180, only wickets matter (economy adds almost nothing)',
        transform=ax.transAxes, fontsize=9, color=TX2)

ax.legend(frameon=True, framealpha=1, edgecolor=DIV,
          fontsize=10, loc='upper right', facecolor=CARD)

plt.tight_layout()
plt.savefig('chart4_bowling_strategy.png', dpi=180, bbox_inches='tight', facecolor=BG)
plt.show()
print('Saved: chart4_bowling_strategy.png')
```

## Table — Top Batter & Bowler Per Season

```python
# ── Compute from df ──
# Top batter per season by total runs
batter_season = df.groupby(['season', 'batter'])['runs_batter'].sum().reset_index()
top_batters = (
    batter_season
    .sort_values('runs_batter', ascending=False)
    .groupby('season')
    .first()
    .reset_index()
    .rename(columns={'batter': 'Top Batter', 'runs_batter': 'Runs'})
    .sort_values('season')
)

# Top bowler per season by wickets
wickets_df = df[df['wicket_kind'] != 'No Wicket']
bowler_season = wickets_df.groupby(['season', 'bowler']).size().reset_index(name='wickets')
top_bowlers = (
    bowler_season
    .sort_values('wickets', ascending=False)
    .groupby('season')
    .first()
    .reset_index()
    .rename(columns={'bowler': 'Top Bowler', 'wickets': 'Wickets'})
    .sort_values('season')
)

# Merge
season_table = top_batters[['season', 'Top Batter', 'Runs']].merge(
    top_bowlers[['season', 'Top Bowler', 'Wickets']], on='season'
)
season_table.columns = ['Season', 'Top Batter', 'Runs', 'Top Bowler', 'Wickets']
print(season_table.to_string(index=False))
```

## Table Visual — Styled Season Leaders

```python
n_rows = len(season_table)
fig_h  = 1.1 + n_rows * 0.42

fig, ax = plt.subplots(figsize=(13, fig_h))
fig.patch.set_facecolor(BG)
ax.set_facecolor(BG)
ax.axis('off')

col_labels = ['Season', 'Top Batter', 'Runs', 'Top Bowler', 'Wickets']
col_widths = [0.10,     0.30,          0.12,   0.30,         0.12]
col_aligns = ['center', 'left',        'center','left',      'center']
col_x      = [0.02, 0.13, 0.44, 0.57, 0.88]

total_w = 0.98
row_h   = 0.042
header_y= 0.97

# Header background
header_rect = FancyBboxPatch((0.01, header_y - row_h*0.8), total_w, row_h*1.1,
                              boxstyle='round,pad=0.005',
                              facecolor=TX, edgecolor='none',
                              transform=ax.transAxes, zorder=2)
ax.add_patch(header_rect)

# Header labels
header_icons = ['', '🏏', '', '🎯', '']
for lbl, icon, cx, align in zip(col_labels, header_icons, col_x, col_aligns):
    ax.text(cx, header_y - row_h*0.15,
            f'{icon}  {lbl}'.strip() if icon else lbl,
            transform=ax.transAxes,
            fontsize=10, fontweight='bold', color='white',
            ha=align, va='center', zorder=3)

# Rows
for i, row in season_table.iterrows():
    row_y  = header_y - row_h * (i + 2)
    row_bg = CARD if i % 2 == 0 else '#F1F3F5'

    row_rect = FancyBboxPatch((0.01, row_y - row_h*0.6), total_w, row_h,
                               boxstyle='round,pad=0.002',
                               facecolor=row_bg, edgecolor='none',
                               transform=ax.transAxes, zorder=1)
    ax.add_patch(row_rect)

    values = [str(row['Season']), row['Top Batter'],
              str(int(row['Runs'])), row['Top Bowler'],
              str(int(row['Wickets']))]

    for val, cx, align in zip(values, col_x, col_aligns):
        is_runs    = (cx == col_x[2])
        is_wickets = (cx == col_x[4])
        color = C1 if is_runs else (C4 if is_wickets else TX)
        fw    = 'bold' if (is_runs or is_wickets) else 'normal'
        ax.text(cx, row_y - row_h*0.08, val,
                transform=ax.transAxes,
                fontsize=10, color=color, fontweight=fw,
                ha=align, va='center', zorder=2)

# Title above table
ax.text(0.01, 1.0, 'Season Leaders — Top Batter & Bowler Each Year',
        transform=ax.transAxes,
        fontsize=14, fontweight='bold', color=TX, va='bottom')
ax.text(0.01, 0.985, 'Most runs scored and most wickets taken per IPL season',
        transform=ax.transAxes,
        fontsize=9, color=TX2, va='bottom')

plt.tight_layout()
plt.savefig('table_season_leaders.png', dpi=180, bbox_inches='tight', facecolor=BG)
plt.show()
print('Saved: table_season_leaders.png')
```
