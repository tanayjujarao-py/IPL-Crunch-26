```python
# Total runs per team per phase per match
phase_runs = df.groupby(['match_id', 'batting_team', 'phase'])['runs_total'].sum().reset_index()

# Merge with match winner
phase_runs = phase_runs.merge(df[['match_id','winner']].drop_duplicates(), on='match_id')
phase_runs['won'] = phase_runs['batting_team'] == phase_runs['winner']

# Average runs by phase for winners vs losers
phase_runs.groupby(['phase', 'won'])['runs_total'].mean()
```

```python
# Count balls per phase per match per team
phase_balls = df.groupby(['match_id', 'batting_team', 'phase'])['ball'].count().reset_index()
phase_balls.rename(columns={'ball': 'balls'}, inplace=True)

# Merge with phase_runs, then compute RPI (runs per ball * 6)
phase_runs = phase_runs.merge(phase_balls, on=['match_id','batting_team','phase'])
phase_runs['run_rate'] = (phase_runs['runs_total'] / phase_runs['balls']) * 6
```

```python
# Step 1 — Phase runs per team per match per innings
phase_innings = df.groupby(['match_id', 'innings', 'batting_team', 'phase'])['runs_total'].sum().reset_index()

# Step 2 — Tag whether batting team won
winners = df[['match_id', 'winner']].drop_duplicates()
phase_innings = phase_innings.merge(winners, on='match_id')
phase_innings['won'] = phase_innings['batting_team'] == phase_innings['winner']

# Step 3 — For each match+phase+innings, who scored more?
phase_pivot = phase_innings.groupby(['match_id', 'innings', 'phase']).apply(
    lambda g: pd.Series({
        'winner_runs': g[g['won'] == True]['runs_total'].values[0] if True in g['won'].values else None,
        'loser_runs': g[g['won'] == False]['runs_total'].values[0] if False in g['won'].values else None,
    })
).reset_index()

phase_pivot['phase_winner_won_match'] = phase_pivot['winner_runs'] > phase_pivot['loser_runs']

# Step 4 — Win rate when dominating each phase, split by innings
result = phase_pivot.groupby(['innings', 'phase'])['phase_winner_won_match'].mean().mul(100).round(1)
print(result)
```

```python
# === VENUE: BATTING FIRST VS CHASING WIN RATE ===
venue = df[['match_id', 'venue', 'winner']].drop_duplicates()
venue = venue[~venue['winner'].isin(['Tie', 'No Result'])]

# Get innings 1 batting team per match
inn1_teams = df[df['innings'] == 1][['match_id', 'batting_team']].drop_duplicates()

venue = venue.merge(inn1_teams, on='match_id')
venue['batting_first_won'] = venue['batting_team'] == venue['winner']

# Win rate per venue + match count
venue_stats = venue.groupby('venue').agg(
    matches=('match_id', 'count'),
    batting_first_win_rate=('batting_first_won', 'mean')
).reset_index()

venue_stats['batting_first_win_rate'] = (venue_stats['batting_first_win_rate'] * 100).round(1)
venue_stats['chasing_win_rate'] = (100 - venue_stats['batting_first_win_rate']).round(1)

# Filter venues with enough matches to be statistically meaningful
venue_stats = venue_stats[venue_stats['matches'] >= 10].sort_values('batting_first_win_rate', ascending=False)

print(venue_stats.to_string())
```

```python
# Do captains make the right toss decision at each venue?
toss = df[['match_id', 'venue', 'toss_winner', 'toss_decision', 'winner']].drop_duplicates()
toss = toss[~toss['winner'].isin(['Tie', 'No Result'])]
toss = toss[toss['venue'].isin(venue_stats['venue'])]

venue_toss = toss.groupby(['venue', 'toss_decision']).size().reset_index(name='count')
venue_toss_pct = venue_toss.groupby('venue').apply(
    lambda g: (g[g['toss_decision']=='field']['count'].sum() / g['count'].sum() * 100)
).round(1).reset_index()
venue_toss_pct.columns = ['venue', 'pct_chose_to_field']

# Merge with venue stats
venue_final = venue_stats.merge(venue_toss_pct, on='venue')
venue_final['correct_decision_pct'] = venue_final.apply(
    lambda r: r['pct_chose_to_field'] if r['chasing_win_rate'] > 50 
              else 100 - r['pct_chose_to_field'], axis=1
).round(1)

print(venue_final[['venue', 'matches', 'batting_first_win_rate', 'chasing_win_rate', 'pct_chose_to_field', 'correct_decision_pct']].to_string())
```

```python
# === DOT BALL PRESSURE ANALYSIS ===
# Step 1: Create a ball sequence per match per innings per batting team
pressure = df[df['innings'].isin([1, 2])].copy()
pressure = pressure.sort_values(['match_id', 'innings', 'over', 'ball'])

# Tag dot balls (no runs scored off the bat, no extras)
pressure['is_dot'] = (pressure['runs_total'] == 0).astype(int)

# Step 2: Count consecutive dots using groupby + cumsum trick
pressure['ball_index'] = pressure.groupby(['match_id', 'innings']).cumcount()

# Rolling window — check if last 3 balls were all dots
pressure['prev1'] = pressure.groupby(['match_id', 'innings'])['is_dot'].shift(1)
pressure['prev2'] = pressure.groupby(['match_id', 'innings'])['is_dot'].shift(2)
pressure['prev3'] = pressure.groupby(['match_id', 'innings'])['is_dot'].shift(3)

# Flag: this delivery comes after 3+ consecutive dots
pressure['after_3_dots'] = (
    (pressure['prev1'] == 1) & 
    (pressure['prev2'] == 1) & 
    (pressure['prev3'] == 1)
)

# Step 3: What happened on the delivery AFTER 3 dots?
after_dots = pressure[pressure['after_3_dots'] == True].copy()

after_dots['outcome'] = 'other'
after_dots.loc[after_dots['runs_batter'] == 0, 'outcome'] = 'dot'
after_dots.loc[after_dots['runs_batter'].isin([4, 6]), 'outcome'] = 'boundary'
after_dots.loc[after_dots['runs_batter'].isin([1, 2, 3]), 'outcome'] = 'running'
after_dots.loc[after_dots['wicket_kind'] != 'No Wicket', 'outcome'] = 'wicket'

print("=== OUTCOME AFTER 3+ DOT BALLS ===")
outcome_dist = after_dots['outcome'].value_counts(normalize=True).mul(100).round(1)
print(outcome_dist)

print(f"\nTotal deliveries after 3+ dots: {len(after_dots)}")

# Step 4: Split by phase
print("\n=== BY PHASE ===")
print(after_dots.groupby('phase')['outcome'].value_counts(normalize=True).mul(100).round(1))
```

```python
# === NORMAL DELIVERY OUTCOMES (no dot pressure) ===
normal = pressure[pressure['after_3_dots'] == False].copy()

normal['outcome'] = 'other'
normal.loc[normal['runs_batter'] == 0, 'outcome'] = 'dot'
normal.loc[normal['runs_batter'].isin([4, 6]), 'outcome'] = 'boundary'
normal.loc[normal['runs_batter'].isin([1, 2, 3]), 'outcome'] = 'running'
normal.loc[normal['wicket_kind'] != 'No Wicket', 'outcome'] = 'wicket'

print("=== NORMAL DELIVERY OUTCOMES ===")
print(normal['outcome'].value_counts(normalize=True).mul(100).round(1))

print("\n=== NORMAL BY PHASE ===")
print(normal.groupby('phase')['outcome'].value_counts(normalize=True).mul(100).round(1))
```

```python
# -----------------------------
# Phase Classification
# -----------------------------
def get_phase(over):
    if over <= 5:
        return 'Powerplay'
    elif over <= 14:
        return 'Middle Overs'
    else:
        return 'Death Overs'

df['phase'] = df['over'].apply(get_phase)

# -----------------------------
# Total runs per innings-phase
# -----------------------------
phase_runs = (
    df.groupby(['match_id', 'innings', 'batting_team', 'phase'])['runs_total']
    .sum()
    .reset_index()
)

# -----------------------------
# Match winner info
# -----------------------------
match_winner = (
    df[['match_id', 'winner']]
    .drop_duplicates()
)

# Merge winner
phase_runs = phase_runs.merge(match_winner, on='match_id', how='left')

# -----------------------------
# Winning / Losing Tag
# -----------------------------
phase_runs['result'] = phase_runs.apply(
    lambda x: 'Winning Team'
    if x['batting_team'] == x['winner']
    else 'Losing Team',
    axis=1
)

# -----------------------------
# Average runs per phase
# -----------------------------
avg_phase_runs = (
    phase_runs.groupby(['phase', 'result'])['runs_total']
    .mean()
    .reset_index()
)

# Round values
avg_phase_runs['runs_total'] = avg_phase_runs['runs_total'].round(2)

print(avg_phase_runs)
```
